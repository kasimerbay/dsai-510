```python
# Run this cell first
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#import scipy.stats as stats
import statsmodels.api as sm

# Set the display option to show all rows scrolling with a slider
pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', None)


# pd.options.display.max_rows = 1000
# pd.options.display.max_columns = 20

# To disable this, run the line below:
# pd.reset_option('display.max_rows')
# pd.reset_option('display.max_columns')
```

```python
advertising = pd.read_csv('./Advertising.csv').drop(columns='Unnamed: 0')

print("Number of duplicate records in 'Advertising':",advertising.duplicated().sum()  )

advertising
```

## Model: sales ~ newspaper

```python
X = advertising[['newspaper']]
X = sm.add_constant(X)
y = advertising['sales']
model = sm.OLS(y, X).fit()

print(model.summary())
```

## Model: sales ~ TV + radio + newspaper

```python
X = advertising[['TV', 'radio', 'newspaper']]
X = sm.add_constant(X)
y = advertising['sales']
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Generate the predicted values
predictions = model.predict(X)

# Calculate the residuals
residuals = y - predictions

# Create a residual plot
plt.scatter(predictions, residuals)
plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max(), colors='red', linestyles='--')
plt.xlabel('Predicted Values (=Fitted Values=Model Values)')
plt.ylabel('Residuals (= y_data - y_model)')
plt.title('Residual Plot')
plt.show()
```

## Model: sales ~ TV + newspaper + radio + radio$^2$

```python
X = advertising[['TV', 'radio', 'newspaper']]
X['radio_squared'] = X['radio'] ** 2
X = sm.add_constant(X)
y = advertising['sales']
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Generate the predicted values
predictions = model.predict(X)

# Calculate the residuals
residuals = y - predictions

# Create a residual plot
plt.scatter(predictions, residuals)
plt.hlines(y=0, xmin=predictions.min(), xmax=predictions.max(), colors='red', linestyles='--')
plt.xlabel('Predicted Values (=Fitted Values=Model Values)')
plt.ylabel('Residuals (= y_data - y_model)')
plt.title('Residual Plot')
plt.show()
```

## Model: sales ~ TV + radio + TV $^2$ + radio $^2$ + TV $^3$ + radio $^3$ + TV $\times$ radio + TV $\times$ radio $^2$ + TV $^2$ $\times$ radio

```python
X = advertising[['TV','radio']]
X['radio_squared'] = X['radio'] ** 2
X['radio_cubed'] = X['radio'] ** 3
X['TV_squared'] = X['TV'] ** 2
X['TV_cubed'] = X['TV'] ** 3
X['TV_x_radio'] = X['TV'] * X['radio']
X['TV_x_radio_squared'] = X['TV'] * X['radio_squared']
X['TV_squared_x_radio'] = X['TV_squared'] * X['radio']
X = sm.add_constant(X)
y = advertising['sales']
model = sm.OLS(y, X).fit()

print(model.summary())
```

## Static 3D Plot for the data and flat model without the interaction term

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Coefficients
beta0 = 2.92
beta_TV = 0.046
beta_radio = 0.188

# Predictors and response variable
X_TV = advertising['TV']
X_radio = advertising['radio']
y = advertising['sales']

# Create a grid over the ranges of TV and radio for plotting
TV_range = np.linspace(X_TV.min(), X_TV.max(), 100)
radio_range = np.linspace(X_radio.min(), X_radio.max(), 100)
TV_grid, radio_grid = np.meshgrid(TV_range, radio_range)

# Compute the predicted sales over the grid
y_pred = beta0 + beta_TV * TV_grid + beta_radio * radio_grid

# Create the figure and the 3D Axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(TV_grid, radio_grid, y_pred, alpha=0.5, rstride=100, cstride=100, color='b', edgecolor='none')

# Scatter plot of the actual data points
ax.scatter(X_TV, X_radio, y, color='r')

# Lines from points to the plane
for (xi, yi, zi) in zip(X_TV, X_radio, y):
    # Here we plot a line from the actual point to the predicted plane
    ax.plot([xi, xi], [yi, yi], [zi, beta0 + beta_TV * xi + beta_radio * yi], 'gray', linewidth=1, linestyle='--')

# Set labels
ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')

# Show plot
plt.show()
```

## Interactive 3D Plot for the data and flat model without the interaction term

```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd

# Assuming 'advertising' is your DataFrame
# Coefficients
beta0 = 2.92
beta_TV = 0.046
beta_radio = 0.188

# Create a grid over the ranges of TV and radio for plotting
tv_range = np.linspace(advertising['TV'].min(), advertising['TV'].max(), 100)
radio_range = np.linspace(advertising['radio'].min(), advertising['radio'].max(), 100)
tv_grid, radio_grid = np.meshgrid(tv_range, radio_range)

# Compute the predicted sales over the grid
y_pred = beta0 + beta_TV * tv_grid + beta_radio * radio_grid

# Create the surface plot
surface = go.Surface(x=tv_grid, y=radio_grid, z=y_pred, colorscale='Viridis', opacity=0.6, name='Prediction Plane')

# Create the scatter plot
scatter = go.Scatter3d(
    x=advertising['TV'],
    y=advertising['radio'],
    z=advertising['sales'],
    mode='markers',
    marker=dict(
        size=3,
        color='red'
    ),
    name='Data Points'
)

# Lines from points to the plane
lines = []
for i in range(len(advertising)):
    lines.append(go.Scatter3d(
        x=[advertising['TV'][i], advertising['TV'][i]],
        y=[advertising['radio'][i], advertising['radio'][i]],
        z=[advertising['sales'][i], beta0 + beta_TV * advertising['TV'][i] + beta_radio * advertising['radio'][i]],
        mode='lines',
        line=dict(color='gray', width=2),
        showlegend=False
    ))

# Layout for the plot
layout = go.Layout(
    title='Sales vs. TV and Radio Advertising Budget',
    scene=dict(
        xaxis=dict(title='TV'),
        yaxis=dict(title='Radio'),
        zaxis=dict(title='Sales'),
    )
)

# Combine everything into one figure
fig = go.Figure(data=[surface, scatter] + lines, layout=layout)

# Show the figure
fig.show()
```

## PCA of the predictors radio and newspaper

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'advertising' is your DataFrame with the 'newspaper' and 'radio' columns

# Extracting the two features for PCA
X = advertising[['radio', 'newspaper']].values

# Standardizing the features before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA to reduce the features to 1 principal component
pca = PCA(n_components=1)
pca.fit(X_scaled)

# The PCA component loadings (coefficients)
loadings = pca.components_

# print the loadings for the first principal component
print("Loadings for the first principal component:")
print(f"Radio: {loadings[0,0]}")
print(f"Newspaper: {loadings[0,1]}")
```

## VIF metric to check (multi)collinearity

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# The independent variables set
X = advertising[['TV', 'radio', 'newspaper']]

# Adding a constant to the model (intercept)
X = add_constant(X)

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# Calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Your t-statistic
t_stat = 3.3

# Degrees of freedom
df = 10

# Create a range of values for t
t_values = np.linspace(-10, 10, 1000)
# Calculate the t-distribution PDF for these values
t_distribution = stats.t.pdf(t_values, df)

# Plot the t-distribution
plt.plot(t_values, t_distribution, label='t-distribution')

# Calculate the critical t value for the given alpha level
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df)

# Fill the right side critical region
plt.fill_between(t_values, t_distribution, where=(t_values >= t_crit), color='red', alpha=0.5)

# Since it's a two-tailed test, also fill the left side critical region
plt.fill_between(t_values, t_distribution, where=(t_values <= -t_crit), color='red', alpha=0.5)

# Annotate the t statistic
plt.axvline(x=t_stat, color='black', linestyle='--', label=f't-statistic = {t_stat}')

# Annotate the critical t value
plt.axvline(x=t_crit, color='green', linestyle='--', label=f'Critical t-value = {round(t_crit,2)}')

# Add labels and legend
plt.xlabel('t value')
plt.ylabel('Probability Density Function')
plt.title('t-Distribution with Shaded p-Value Area')
plt.legend()

# Show the plot
plt.show()

# Calculate the two-tailed p-value
p_value = 2 * (stats.t.sf(abs(t_stat), df))
print(f'The p-value for a t-statistic of {t_stat} with {df} degrees of freedom is: {p_value}')
```
